```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\varsh\Downloads\Adidas US Sales Datasets 1.csv")
df
```

```python
df.info()
df.describe()
```

```python
sales_by_city = df.groupby("City")["Total Sales"].sum().reset_index()
sales_by_city = sales_by_city.sort_values(by="Total Sales", ascending=False)
print(sales_by_city.head(10))  
```

```python
sales_by_state = df.groupby("State")["Total Sales"].sum().reset_index()
sales_by_state = sales_by_state.sort_values(by="Total Sales", ascending=False)
print(sales_by_state.head(10))  
```

```python
sales_by_region = df.groupby("Region")["Total Sales"].sum().reset_index()
sales_by_region = sales_by_region.sort_values(by="Total Sales", ascending=False)
print(sales_by_region)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=sales_by_city["City"][:10], y=sales_by_city["Total Sales"][:10], palette="Blues_r")
plt.xticks(rotation=45)
plt.title("Top 10 Cities by Sales")
plt.xlabel("City")
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=sales_by_region["Region"], y=sales_by_region["Total Sales"], palette="coolwarm")
plt.title("Sales Performance by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
best_city = sales_by_city.iloc[0]  
worst_city = sales_by_city.iloc[-1]
print(f"Best Performing City: {best_city['City']} - Sales: ${best_city['Total Sales']}")
print(f"Worst Performing City: {worst_city['City']} - Sales: ${worst_city['Total Sales']}")
```

```python
sales_by_retailer = df.groupby("Retailer")["Total Sales"].sum().reset_index()
sales_by_retailer = sales_by_retailer.sort_values(by="Total Sales", ascending=False)
sales_by_retailer.head(3) 
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x=df["Price per Unit"], y=df["Total Sales"])
plt.title("Price per Unit vs Total Sales")
plt.xlabel("Price per Unit")
plt.ylabel("Total Sales")
plt.show()
```

```python
sales_by_product = df.groupby("Product")["Total Sales"].sum().reset_index()
sales_by_product = sales_by_product.sort_values(by="Total Sales", ascending=False)
sales_by_product.head(3)
```

```python
plt.figure(figsize=(6,6))
sns.barplot(x=sales_by_product["Product"][:10], y=sales_by_product["Total Sales"][:10], palette="coolwarm")
plt.xticks(rotation=45)
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

```python
sales_by_method = df.groupby("Sales Method")["Total Sales"].sum().reset_index()
sales_by_method = sales_by_method.sort_values(by="Total Sales", ascending=False)
sales_by_method
```

```python
plt.figure(figsize=(6,6))
sns.barplot(x=sales_by_method["Sales Method"], y=sales_by_method["Total Sales"], palette="coolwarm")
plt.xticks(rotation=45)
plt.xlabel("Sales Method")
plt.ylabel("Total Sales")
plt.show()
```

```python
popular_products = df.groupby("Product")["Units Sold"].sum().reset_index()
popular_products = popular_products.sort_values(by="Units Sold", ascending=False)
popular_products
```

```python
popular_products.plot(kind = 'bar')
sns.barplot(x=popular_products["Product"], y=popular_products["Units Sold"], palette="coolwarm")
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x=df["Price per Unit"], y=df["Units Sold"])
plt.title("Price per Unit vs Units Sold")
plt.xlabel("Price per Unit")
plt.ylabel("Units Sold")
plt.show()
```

```python
df["Invoice Date"] = pd.to_datetime(df["Invoice Date"], format="%d-%m-%Y") 
df["Month"] = df["Invoice Date"].dt.month  # Extract month  

sales_by_month = df.groupby("Month")["Units Sold"].sum().reset_index()

plt.figure(figsize=(6,6))
sns.lineplot(x=sales_by_month["Month"], y=sales_by_month["Units Sold"], marker="o", color="b")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Units Sold")
plt.xticks(range(1,13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.show()
```
